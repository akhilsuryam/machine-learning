```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture

# Load the data
df = pd.read_csv('data2.csv')
```

```python
# Set the number of clusters
k = 2

# Initialize the KMeans model
kmeans = KMeans(n_clusters=k)

# Fit the model to the data
kmeans.fit(df)

# Assign the clusters to the data
df['kmeans_cluster'] = kmeans.predict(df)

```

```python
def print_clusters(df):
  for cluster, data in df.groupby('kmeans_cluster'):
    print(f'Cluster {cluster}:')
    print(data)

print_clusters(df)
```

```python
def print_clusters(df):
  for cluster, data in df.groupby('kmeans_cluster'):
    print(f'Cluster {cluster}:')
    print(data)

print_clusters(df)
```

```python
# Initialize the DBSCAN model
dbscan = DBSCAN()

# Fit the model to the data
dbscan.fit(df)

# Assign the clusters to the data
df['dbscan_cluster'] = dbscan.labels_

```

```python
def print_clusters(df):
  for cluster, data in df.groupby('dbscan_cluster'):
    print(f'Cluster {cluster}:')
    print(data)

print_clusters(df)
```

```python
# Set the number of clusters
k = 5

# Initialize the AgglomerativeClustering model
agg = AgglomerativeClustering(n_clusters=k)

# Fit the model to the data
agg.fit(df)

# Assign the clusters to the data
df['agg_cluster'] = agg.labels_
```

```python
# Set the number of clusters
k = 5

# Initialize the GaussianMixture model
gm = GaussianMixture(n_components=k)

# Fit the model to the data
gm.fit(df)

# Assign the clusters to the data
df['gm_cluster'] = gm.predict(df)
```

```python
def print_clusters(df, column):
  for cluster, data in df.groupby(column):
    print(f'Cluster {cluster}:')
    print(data)
```

```python

print('KMeans Clusters:')
print_clusters(df, 'kmeans_cluster')
```

```python

print('\nDBSCAN Clusters:')
print_clusters(df, 'dbscan_cluster')
```

```python

print('\nAgglomerativeClustering Clusters:')
print_clusters(df, 'agg_cluster')
```

```python

print('\nGaussianMixture Clusters:')
print_clusters(df, 'gm_cluster')
```
